**5. Selecting Input and Output Variables**

```python
feature_cols = [
    'disposedAmount',
    'pendingAmount',
    'pendingCourt',
    'disposedCourt',
    'totalCourt',
    'month',
    'day_of_week',
    'is_weekend'
]

target = "pendingChallan"

X = df[feature_cols]
y = df[target]
```

**6. Train / Test Split Data**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#importing library
from sklearn.model_selection import train_test_split

#Splitting training and testing data
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.20, random_state=42, shuffle=True)

print("Model splitted successfully... ")
```

**7. Applying Random Forest Regressor Algorithm and Train the data with it***

```python
#Importing RandomForestRegressor library
from sklearn.ensemble import RandomForestRegressor

#Train data with the RFR algorithm
rf = RandomForestRegressor(
    n_estimators = 300,
    max_depth = 12,
    random_state = 42
)

rf.fit(Xtrain, ytrain)
rf_pred = rf.predict(Xtest)
```

**Linear Regressor**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(Xtrain, ytrain)
lr_pred = lr.predict(Xtest)
```

**Decision Tree Regressor**

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state=42)
dt.fit(Xtrain, ytrain)
dt_pred = dt.predict(Xtest)
```

**8. Evaluating the model**

```python
#importing libraries related to evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

#creating pred variable as the predicted output of the model
rf_pred = rf.predict(Xtest)

#calculating different evaluation metrics regarding pred
mae = mean_absolute_error(ytest,rf_pred)
mse = np.sqrt(mean_squared_error(ytest,rf_pred))
r2 = r2_score(ytest, rf_pred)

#Displaying metric outputs
print("MAE : ",mae)
print("MSE : ",mse)
print("R² Score : ",r2)
```

```python


def evaluate(name, ytest, pred):
    print(f"\n====== {name} ======")
    print("MAE:", mean_absolute_error(ytest, pred))
    print("RMSE:", np.sqrt(mean_squared_error(ytest, pred)))
    print("R2 Score:", r2_score(ytest, pred))

evaluate("Linear Regression", ytest, lr_pred)
evaluate("Decision Tree", ytest, dt_pred)
evaluate("Random Forest", ytest, rf_pred)
```

```python
def evaluate_train_test(name, model, Xtrain, ytrain, Xtest, ytest):
    train_r2 = model.score(Xtrain, ytrain)
    test_r2 = model.score(Xtest, ytest)

    print(f"\n===== {name} =====")
    print(f"Train R2 Score: {train_r2}")
    print(f"Test R2 Score:  {test_r2}")

    if train_r2 - test_r2 > 0.05:
        print("⚠ Overfitting detected (Train >> Test)")
    elif test_r2 - train_r2 > 0.05:
        print("⚠ Underfitting detected (Test >> Train)")
    else:
        print("✔ Model generalizes well.")
```

```python
evaluate_train_test("Linear Regression", lr, Xtrain, ytrain, Xtest, ytest)
evaluate_train_test("Decision Tree", dt, Xtrain, ytrain, Xtest, ytest)
evaluate_train_test("Random Forest", rf, Xtrain, ytrain, Xtest, ytest)
```

**Building UI using Streamlit**

```python
import joblib
import streamlit as st
import numpy as np
import datetime

# Load model
model = joblib.load("pending_model.pkl")

st.title("🚦 Pending Challan Prediction System (SDG 16)")
st.write("Enter the following details to predict pending challans.")

# INPUTS (Must match model features)
pendingAmount = st.number_input("Pending Amount (₹)", min_value=0)
disposedAmount = st.number_input("Disposed Amount (₹)", min_value=0)

pendingCourt = st.number_input("Pending Court Cases", min_value=0)
disposedCourt = st.number_input("Disposed Court Cases", min_value=0)
totalCourt = st.number_input("Total Court Cases", min_value=0)

# Date Input
date = st.date_input("Select Date", datetime.date.today())

# Extract date-based features
month = date.month
day_of_week = date.weekday()
is_weekend = 1 if day_of_week >= 5 else 0

if st.button("Predict Pending Challans"):
    input_data = np.array([[disposedAmount, pendingAmount,
                            pendingCourt, disposedCourt, totalCourt,
                            month, day_of_week, is_weekend]])

    prediction = int(model.predict(input_data)[0])

    st.success(f"📌 Predicted Pending Challans: **{prediction}**")
```
